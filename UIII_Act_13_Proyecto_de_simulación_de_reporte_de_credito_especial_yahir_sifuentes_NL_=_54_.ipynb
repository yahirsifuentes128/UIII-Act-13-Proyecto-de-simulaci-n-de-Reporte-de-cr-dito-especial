```python
# --- SISTEMA DE REPORTE DE CRÉDITO ESPECIAL CON IA ---

class ReporteCreditoEspecial:
    """Clase que representa el historial financiero de un cliente"""
    def __init__(self, nombre, edad, puntaje, deuda, estado):
        # Atributos que definen el estado del objeto [3, 6]
        self.nombre = nombre
        self.edad = int(edad)
        self.puntaje = int(puntaje)
        self.deuda = float(deuda)
        self.estado = estado

    def analizar_credito_ia(self):
        """Función de IA que evalúa el historial y recomienda una acción [5]"""
        # Lógica basada en operadores relacionales y lógicos [7]
        if self.puntaje >= 700 and self.edad >= 18 and self.deuda < 50000:
            return "IA: Perfil Excelente. Recomendación: APROBAR de inmediato."
        elif self.puntaje >= 500 and self.deuda < 100000:
            return "IA: Riesgo Moderado. Recomendación: REVISIÓN detallada."
        else:
            return "IA: Riesgo Alto. Recomendación: RECHAZAR crédito."

# Lista global para almacenar los objetos de tipo Reporte [8]
base_datos_reportes = []

def agregar_reporte():
    """Registra un nuevo cliente en el sistema [9]"""
    print("\n--- REGISTRO DE NUEVO REPORTE ---")
    nom = input("Nombre del cliente: ")
    ed = input("Edad: ")
    pun = input("Puntaje de crédito (0-1000): ")
    deu = input("Deuda actual ($): ")
    est = input("Estado inicial (Aprobado/Pendiente/Rechazado): ")

    # Instanciación del objeto [4, 10]
    nuevo = ReporteCreditoEspecial(nom, ed, pun, deu, est)
    base_datos_reportes.append(nuevo)
    print(f"¡Reporte de {nom} agregado con éxito!")

def mostrar_reportes():
    """Muestra la lista de clientes en formato de tabla alineada [11, 12]"""
    if not base_datos_reportes:
        print("\nNo hay reportes registrados.")
    else:
        print("\n" + "="*125)
        # Encabezados con f-strings para alineación profesional [13]
        print(f"{'CLIENTE':<20} | {'EDAD':<5} | {'PUNTAJE':<8} | {'DEUDA ($)':<12} | {'ESTADO':<12} | {'ANÁLISIS IA'}")
        print("-" * 125)
        for r in base_datos_reportes:
            # Llamada al método de IA para cada objeto [14, 15]
            ia_resultado = r.analizar_credito_ia()
            print(f"{r.nombre:<20} | {r.edad:<5} | {r.puntaje:<8} | {r.deuda:<12,.2f} | {r.estado:<12} | {ia_resultado}")
        print("="*125)

def actualizar_reporte():
    """Modifica datos financieros de un cliente existente [16, 17]"""
    buscado = input("\nIngrese el nombre del cliente a actualizar: ")
    for r in base_datos_reportes:
        if r.nombre.lower() == buscado.lower():
            print(f"Editando reporte de {r.nombre}...")
            r.puntaje = int(input("Nuevo Puntaje: "))
            r.deuda = float(input("Nueva Deuda ($): "))
            r.estado = input("Nuevo Estado del Crédito: ")
            print("Datos actualizados correctamente.")
            return
    print("Cliente no localizado.")

def eliminar_reporte():
    """Borra un registro de la base de datos [16, 18]"""
    buscado = input("\nNombre del cliente a eliminar: ")
    for i in range(len(base_datos_reportes)):
        if base_datos_reportes[i].nombre.lower() == buscado.lower():
            base_datos_reportes.pop(i)
            print(f"El reporte de {buscado} ha sido eliminado.")
            return
    print("No se encontró el registro.")

# --- MENÚ INTERACTIVO PRINCIPAL ---
def ejecutar_sistema():
    """Controlador del flujo del programa mediante un ciclo interactivo [19, 20]"""
    while True:
        print("\n--- MENÚ: REPORTE DE CRÉDITO ESPECIAL IA ---")
        print("1. Agregar Reporte | 2. Mostrar Reportes | 3. Actualizar | 4. Eliminar | 5. Salir")
        opcion = input("Seleccione una opción (1-5): ")

        if opcion == "1":
            agregar_reporte()
        elif opcion == "2":
            mostrar_reportes()
        elif opcion == "3":
            actualizar_reporte()
        elif opcion == "4":
            eliminar_reporte()
        elif opcion == "5":
            print("Cerrando sistema financiero... programa realizado por yahir sifuentes NL = 54")
            break
        else:
            print("Opción inválida, intente de nuevo.")

# Inicio del programa [21]
if __name__ == "__main__":
    ejecutar_sistema()
```
